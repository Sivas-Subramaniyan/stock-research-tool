```python
#%%
import pandas as pd

# ---------- CONFIG ----------
INPUT_FILE = "screener_results.xlsx"   # your original Screener export
OUTPUT_FILE = "screener_results_cleaned.xlsx"  # cleaned copy
SHEET = 0                              # sheet name or index

# ---------- READ ----------
df = pd.read_excel(INPUT_FILE, sheet_name=SHEET)

print("ðŸ“„ Original column names:")
for i, col in enumerate(df.columns):
    print(f"{i+1:02d}. {repr(col)}")

# ---------- CLEAN HEADERS ----------
def clean_column_name(col):
    """Clean a single column name: remove non-breaking spaces, trim spaces."""
    return (
        str(col)
        .replace('\xa0', ' ')     # replace non-breaking spaces
        .replace('\u200b', '')    # remove zero-width spaces (sometimes hidden)
        .strip()                  # remove leading/trailing whitespace
    )

df.columns = [clean_column_name(c) for c in df.columns]

# ---------- SAVE ----------
with pd.ExcelWriter(OUTPUT_FILE, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name="Cleaned", index=False)

print("\nâœ… Cleaned column names:")
for i, col in enumerate(df.columns):
    print(f"{i+1:02d}. {repr(col)}")

print(f"\nâœ¨ Cleaned Excel saved as: {OUTPUT_FILE}")
```
